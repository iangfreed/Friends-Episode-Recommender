```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_distances
from sklearn.metrics.pairwise import manhattan_distances
from sklearn.metrics.pairwise import euclidean_distances
```

```python
pd.options.display.max_columns = 40
pd.options.display.max_colwidth = 70
```

```python
with open('friends_df.pkl','rb') as picklefile:
    df = pickle.load(picklefile)
```

```python
columns = [f"Topic {i}" for i in range(30)]
```

```python
df_topics = df.iloc[:,5:].copy()
```

```python
df_topics = df_topics.set_index(df['episode_name'])
```

```python
df_topics
```

```python
cos_dists = cosine_distances(df_topics)
man_dists = manhattan_distances(df_topics)
euc_dists = euclidean_distances(df_topics)
```

```python
df_cos = pd.DataFrame(cos_dists,index=df['episode_name'],columns=df['episode_name'])
df_cos
```

```python
df_man = pd.DataFrame(man_dists,index=df['episode_name'],columns=df['episode_name'])
df_man
```

```python
df_euc = pd.DataFrame(euc_dists,index=df['episode_name'],columns=df['episode_name'])
df_euc
```

```python
plt.figure(num=None, figsize=(20, 14)) 
sns.heatmap(df_cos)
```

```python
def top_5_recommended(episode,dist_metric='cosine'):
    if dist_metric == 'cosine':
        return df_cos.loc[episode].sort_values()[1:6].index.values
    elif dist_metric == 'manhattan':
        return df_man.loc[episode].sort_values()[1:6].index.values
    elif dist_metric == 'euclidean':
        return df_euc.loc[episode].sort_values()[1:6].index.values
```

```python
def top_5_recommended(episode,dist_metric='cosine'):
    if dist_metric == 'cosine':
        eps = df_cos.loc[episode].sort_values()[1:6].index.values
        script_links = []
        net_links = []
        for episode in eps:
            script_links.append(df[df['episode_name']==episode]['script_link'].values[0])
            net_links.append(df[df['episode_name']==episode]['netflix_link'].values[0])
        return list(zip(eps,net_links,script_links))
    elif dist_metric == 'manhattan':
        eps = df_man.loc[episode].sort_values()[1:6].index.values
        script_links = []
        net_links = []
        for episode in eps:
            script_links.append(df[df['episode_name']==episode]['script_link'].values[0])
            net_links.append(df[df['episode_name']==episode]['netflix_link'].values[0])
        return list(zip(eps,net_links,script_links))
    elif dist_metric == 'euclidean':
        eps = df_euc.loc[episode].sort_values()[1:6].index.values
        script_links = []
        net_links = []
        for episode in eps:
            script_links.append(df[df['episode_name']==episode]['script_link'].values[0])
            net_links.append(df[df['episode_name']==episode]['netflix_link'].values[0])
        return list(zip(eps,net_links,script_links))
```

```python
top_5_recommended('Episode 514: The One Where Everyone Finds Out','cosine')
```
