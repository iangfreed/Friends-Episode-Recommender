### Imports and Options

```python
import pickle
import pandas as pd
import gensim
from nltk.corpus import stopwords
```

```python
pd.options.display.max_columns = 30
pd.options.display.max_rows = 230
```

```python
with open('friends_df.pkl','rb') as picklefile:
    df = pickle.load(picklefile)
```

```python
df
```

```python
stop_words = stopwords.words('english')
stop_words += ['ineapig','additional','transcribing','by','eric','aasen','(note:','previously','unseen','parts','shown','blue','text.)','[scene:']
```

```python
scripts = list(df['script'])
```

```python
text = [[word for word in script.lower().split() if word not in stop_words] for script in scripts]
```

```python
model = gensim.models.Word2Vec(text,size=100, window=5, min_count=1, workers=2, sg=1)
```

```python
list(model.wv.vocab.items())[:50]
```

```python
model.most_similar("rachel",topn=10)
```

```python
model.similarity('central','perk')
```

```python
model.most_similar('smelly')
```

```python
model.doesnt_match('ross rachel chandler monica phoebe joey'.split())
```
